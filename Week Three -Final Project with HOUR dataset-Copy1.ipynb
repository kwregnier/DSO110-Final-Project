```python
# Remove outliers using Z score method
upper_limit = hour1['cnt'].mean()+3*hour1['cnt'].std()
lower_limit = hour1['cnt'].mean()-3*hour1['cnt'].std()
print('upper_limit', upper_limit)
print('lower_limit', lower_limit)
```

```python
# Find the outliers
hour1.loc[(hour1['cnt']> upper_limit) | (hour1['cnt']< lower_limit)]
```

```python
# Trim or delete outliers
new_hour1 = hour1.loc[(hour1['cnt']< upper_limit) & (hour1['cnt']> lower_limit)]
print('before removing outliers:', len(hour1))
print('after removing outliers:', len(new_hour1))
print('outliers:', len(hour1) - len(new_hour1))
```

```python
# 244 outliers have been removed, let's see the new_hour1 boxplot
sns.boxplot(new_hour1['cnt'])
```
